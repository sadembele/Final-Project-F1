```python
import pandas as pd
```

# Load Data

```python
pitstops = pd.read_csv("PitStopFinal.csv")
```

```python
pitstops
```

# Create an index for values greater than or equal to 995856 (pit times we do not want)

```python
indexNames = pitstops[ pitstops['milliseconds'] >= 995856 ].index
```

# Drop Indexed Rows

```python
pitstops.drop(indexNames , inplace=True)
```

# Save changes to CSV file

```python
pitstops.to_csv("updated_PitStopFinal.csv",index=False)
```
